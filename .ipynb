```python
import pandas as pd
import os
import numpy as np
```

```python
#file to load
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"
clean_student_data_to_load = "clean_students_complete.csv"
```

```python
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()
```

```python
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
student_names = student_data_df["student_name"].tolist()
#student_names
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
```

```python
#student_names = student_data_df["student_name"].tolist()
#for name in student_names:
#    print(name.split(), len(name.split())) 
```

```python
#Replace the reading and math scores for ninth graders at Thomas High School with NaN
student_data_df.loc[( student_data_df ["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th") & (student_data_df["reading_score"] > 0) , "reading_score"] = np.nan
student_data_df.loc[( student_data_df ["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th") & (student_data_df["math_score"] > 0) , "math_score"] = np.nan
#student_data_df.head()
```

```python
school_data_nan_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_nan_df.head()
```

```python
#Merge the clean student data with the school dataset
clean_student_data_df = pd.read_csv(clean_student_data_to_load)
school_data_complete_df = pd.merge(clean_student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
school_count = len(school_data_complete_df ["school_name"].unique())
student_count = student_data_df["student_name"].count()
student_count_per_school=school_data_df.set_index(["school_name"])["size"]
school_types = school_data_df.set_index(["school_name"])["type"]
per_school_budget =  school_data_df.set_index(["school_name"])["budget"]
total_budget=per_school_budget.sum()
per_school_capita=per_school_budget/student_count_per_school

average_reading_score = school_data_complete_df["reading_score"].mean()
average_math_score = school_data_complete_df["math_score"].mean()
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]

average_reading_score_nan = school_data_nan_df["reading_score"].mean()
average_math_score_nan = school_data_nan_df["math_score"].mean()
passing_math_nan = school_data_nan_df[school_data_nan_df["math_score"] >= 70]
passing_reading_nan = school_data_nan_df[school_data_nan_df["reading_score"] >= 70]
```

```python
# Calculate the number of students passing math.
passing_math_count = passing_math["student_name"].count()
passing_math_count_nan =  passing_math_nan["student_name"].count()

# Calculate the number of students passing reading.
passing_reading_count = passing_reading["student_name"].count()
passing_reading_count_nan = passing_reading_nan["student_name"].count()
```

```python
# Calculate the percent that passed math.
passing_math_percentage = passing_math_count / student_count * 100
passing_math_percentage_nan = passing_math_count_nan / student_count * 100

# Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / student_count * 100
passing_reading_percentage_nan = passing_reading_count_nan / student_count * 100
```

```python
overall_passing_percentage = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)&(school_data_complete_df["reading_score"] >= 70)]
overall_passing_percentage = overall_passing_percentage.count()["student_name"]/ student_count * 100
#overall_passing_percentage
```

```python
overall_passing_percentage_nan = school_data_nan_df[(school_data_nan_df["math_score"] >= 70)&(school_data_nan_df["reading_score"] >= 70)]
overall_passing_percentage_nan = overall_passing_percentage_nan.count()["student_name"]/ student_count * 100
#overall_passing_percentage_nan 
```

```python
district_summary_df = pd.DataFrame([{
    "Row": "Original"  ,      
    "Total Schools": len(school_types),
             "Total Students": student_count,
             "Total Budget": total_budget,
             "Average Math Score": average_math_score,
           "Average Reading Score": average_reading_score,
           "% Passing Math": passing_math_percentage,
           "% Passing Reading": passing_reading_percentage,
           "% Overall Passing": overall_passing_percentage}])
#district_summary_df
```

```python
district_summary_df_nan = pd.DataFrame([{
    "Row": "After Removal"  ,  
             "Total Schools": len(school_types),
             "Total Students": student_count,
             "Total Budget": total_budget,
             "Average Math Score": average_math_score_nan,
           "Average Reading Score": average_reading_score_nan,
           "% Passing Math": passing_math_percentage_nan,
           "% Passing Reading": passing_reading_percentage_nan,
           "% Overall Passing": overall_passing_percentage_nan}])
#district_summary_df_nan
district_summary=pd.concat([district_summary_df, district_summary_df_nan])
district_summary
```
